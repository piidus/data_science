### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
    Also add title, xlabel, and ylabel to the plot.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.xlabel('X Lavel')
plt.ylabel('Y lavel')
plt.title("An Imaginary Graph")
plt.scatter(x= x, y= y, alpha = 0.7)
plt.show()
```

### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

```python
import numpy as np
ax = np.array([0, 1, 2, 3, 4, 5])
ay = np.array([0, 100, 200, 300, 400, 500])

bx = np.array([0, 1, 2, 3, 4, 5])
by = np.array([50, 20, 40, 20, 60, 70])

cx = np.array([0, 1, 2, 3, 4, 5])
cy = np.array([10, 20, 30, 40, 50, 60])

dx = np.array([0, 1, 2, 3, 4, 5])
dy = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(5,1,1)
plt.plot(ax, ay)
plt.plot(bx, by)
plt.plot(cx, cy)
plt.plot(dx, dy)
```

### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

    A bar chart or bar graph is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. 

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.rcdefaults()
fig, ax = plt.subplots()


y_pos = np.arange(len(company))
performance = 3 + 10 * np.random.rand(len(profit))
error = np.random.rand(len(profit))

ax.barh(y_pos, performance, xerr=error, align='center')
ax.set_yticks(y_pos, labels=company)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Performance')
ax.set_title('Stock Performance')

plt.show()
```

### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

> A box plot which is also known as a whisker plot displays a summary of a set of data containing the minimum, first quartile, median, third quartile, and maximum.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1, box2, box1, box2]
 
fig = plt.figure(figsize =(10, 7))
 
# Creating axes instance
ax = fig.add_axes([0, 1, 1, 1])
 
# Creating plot
bp = ax.boxplot(data)
 
# show plot
plt.show()
```
