###  Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

#### Type Of Charts/Graphs/Plots :
        Line Plot
        Bar Chart
        Pairplot
        Scatter Plot
        Histogram

```python
import seaborn as sns
iris = sns.load_dataset('iris')
iris.head()
```

```python
# Line Plot
sns.lineplot(x = iris['sepal_length'], y = iris['sepal_width'], hue=iris['petal_length'])
```

```python
#  Barchart
sns.barplot(data= iris)
```

```python
# Pairplot
sns.pairplot(iris)
```

```python
# Scatterplot
sns.scatterplot(x = iris['sepal_length'], y = iris['petal_width'], hue=iris['petal_length'])
```

```python
# histogram
sns.histplot(data = iris)
```

### Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

```python
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
sns.scatterplot(data = fmri, x = fmri['timepoint'], y = fmri['signal'], hue = fmri['event'], style = fmri['region'])
```

### Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flight = sns.load_dataset('flights')
flight.head()
```

```python
flight1 = flight.drop(labels='month', axis=1)
flight1.head()
```

```python
sns.heatmap(data = flight1, x = flight1.year, y = flight1.passengers)
```
